```python
import pandas as pd
import numpy as np
import math

def calculateLastWork(str_):
    arr = str_.split(',')
    return arr.count('+')*2.5 + arr.count('+-')*2 + arr.count('-+')

df = pd.read_csv('group.tsv', sep='\t')

data = df.drop(columns="01.10")
df["rating"] = 0
rating = [0]*df.index.size
for i in range (df.index.size):
    rating[i] = rating[i] + list(data.iloc[i].values).count('+') * 2
    if (df.iloc[i]['rating'] + list(df.iloc[i].values).count('н') <= 2):
        rating[i] = rating[i] + 3
    rating[i] = rating[i] + list(data.iloc[i].values).count('-') * -1
    rating[i] = rating[i] + list(data.iloc[i].values).count('+-')
    rating[i] = rating[i] + list(data.iloc[i].values).count('1/2')
    rating[i] = rating[i] + calculateLastWork(df.iloc[i]['01.10'])
df["rating"] = rating 
df["rating20"] = df["rating"].apply(lambda x: math.ceil(x) if x<20 else 20)

df
```
